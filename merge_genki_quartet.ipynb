```python
import pandas as pd
import re
```

```python
# cleaning genki vocab
pdf = pd.read_csv("/content/genki_vocab.csv", header=None)
pdf.columns = ["jp","kana","en", "lesson"]

arr = []
arr2 = []
for l in pdf["lesson"]:
  l = l.replace("読L", "").replace("会L", "").replace("会G", "0")
  l = l.replace("(", "z").replace(",", "z").replace("-", "z")
  l = l.split("z")[0]
  arr.append(int(l))

  if int(l) <= 12:
    arr2.append("genki 1")
  else:
    arr2.append("genki 2")

pdf["lesson"] = arr
pdf["book"] = arr2

pdf = pdf.sort_values(by=['lesson'])
pdf = pdf.drop_duplicates(subset='jp', keep="first")

pdf.head(5)
```

```python
df = pd.DataFrame(columns = ["number","jp","kana","en","lesson", "book"])
#df.columns = ["number","jp","kana","en","book"]

# quartet 1
file_name = "quartet1_vocabularylist_English.xlsx"
xls = pd.ExcelFile(file_name)

for sheet_name in xls.sheet_names:
  sheet = pd.read_excel(file_name, sheet_name=sheet_name, header=None)
  sheet.columns = ["number","jp","kana","en"]
  sheet["lesson"] = [sheet_name.split("-")[0] for i in range(len(sheet))]
  sheet["book"] = ["quartet 1" for i in range(len(sheet))]
  df = pd.concat([df, sheet])

# quartet 2
file_name = "quartet2_vocabularylist_English.xlsx"
xls = pd.ExcelFile(file_name)

for sheet_name in xls.sheet_names:
  sheet = pd.read_excel(file_name, sheet_name=sheet_name, header=None)
  sheet.columns = ["number","jp","kana","en"]
  sheet["lesson"] = [sheet_name.split("-")[0] for i in range(len(sheet))]
  sheet["book"] = ["quartet 2" for i in range(len(sheet))]
  df = pd.concat([df, sheet])

df = df[df["en"].notnull()]
df = df[["jp", "kana", "en", "lesson", "book"]]

# cleaning jp and kana columns to remove brackets etc
arr1 = []
arr2 = []
for i in range(len(df)):
  s1 = df.iloc[i]["jp"]
  s1 = re.sub(r'［.+?］', '', s1)
  s1 = re.sub(r'（.+?）', '', s1)
  arr1.append(s1)

  s2 = df.iloc[i]["kana"]
  s2 = re.sub(r'［.+?］', '', s2)
  s2 = re.sub(r'（.+?）', '', s2)
  arr2.append(s2)

df["jp"] = arr1
df["kana"] = arr2

df = df.drop_duplicates(subset='jp', keep="first")

df.head(5)
```

```python
df = pd.concat([pdf, df])
df = df.drop_duplicates(subset='jp', keep="first")

df.head(5)
```

```python
df.to_csv("genki_quartet_vocab.csv")
```
